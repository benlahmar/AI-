## 1. Rappel et Bases : D√©finition de Fonction

Les fonctions permettent d'encapsuler et de r√©utiliser des blocs de code.

### ‚úèÔ∏è Exercice 1.1 (Fonction Simple)

2.  Elle doit retourner la valeur divis√©e par 255 (une op√©ration courante en traitement d'image pour l'IA).

## 2. Flexibilit√© : Arguments Arbitraires (`*args`)

Le symbole `*` avant un param√®tre dans la d√©finition d'une fonction permet √† celle-ci d'accepter un **nombre ind√©fini d'arguments positionnels**. Ces arguments sont re√ßus sous forme de **tuple**.

### ‚úèÔ∏è Exercice 2.1 (Somme Arbitraire)

1.  D√©finissez une fonction nomm√©e `calculer_moyenne` qui accepte un nombre arbitraire d'arguments (`*nombres`).
2.  Dans la fonction, calculez et retournez la moyenne de ces nombres (somme des √©l√©ments divis√©e par le nombre d'√©l√©ments).
3.  Testez la fonction en lui passant 5 nombres : `calculer_moyenne(10, 20, 30, 40, 50)`.

````python
def calculer_moyenne(*nombres):
    # nombres est un tuple. Utilisez sum() et len()
    if not nombres:
        return 0
    return sum(nombres) / len(nombres)

# Test 1
print(f"Moyenne 1 : {calculer_moyenne(10, 20, 30, 40, 50)}")

# Test 2 avec moins d'arguments
print(f"Moyenne 2 : {calculer_moyenne(5, 15, 25)}")
````

## 3. Concision : Les Expressions Lambda

Une **Expression Lambda** est une petite fonction anonyme (sans nom) d√©finie avec le mot-cl√© `lambda`. Elle est limit√©e √† une seule expression et son r√©sultat est retourn√© implicitement. Elle est souvent utilis√©e pour les fonctions pass√©es en argument.

### Syntaxe :
```python
nom_lambda = lambda arguments: expression
```

### ‚úèÔ∏è Exercice 3.1 (Lambda simple)

1.  Cr√©ez une fonction lambda nomm√©e `carr√©` qui prend un argument `x` et retourne son carr√©.
2.  Testez-la avec la valeur 7.

### ‚úèÔ∏è Exercice 3.2 (Lambda avec deux arguments)

1.  Cr√©ez une fonction lambda nomm√©e `somme_et_multi` qui prend deux arguments `a` et `b` et retourne leur somme multipli√©e par 10.
2.  Testez-la avec 5 et 3.

````python
# 1. Lambda Carr√©
carr√© = lambda x: x ** 2
print(f"Carr√© de 7 : {carr√©(7)}")

# 2. Lambda Somme et Multiplication
somme_et_multi = lambda a, b: (a + b) * 10
print(f"R√©sultat (5, 3) : {somme_et_multi(5, 3)}")
````

## 4. Programmation Fonctionnelle : `map`, `filter`, `reduce`

Ces fonctions de haut niveau permettent d'appliquer des op√©rations √† des collections de mani√®re concise, souvent en utilisant des expressions lambda. 

### A. `map()` : Application √† tous les √©l√©ments

`map(fonction, s√©quence)` applique la fonction √† **chaque** √©l√©ment de la s√©quence.

### ‚úèÔ∏è Exercice 4.1 (Map)

1.  Utilisez `map()` et une expression **lambda** pour normaliser chaque √©l√©ment de la liste `pixels` (diviser par 255.0).
2.  Convertissez le r√©sultat de `map` en une liste pour l'afficher.

````python
pixels = [10, 50, 150, 200, 255]

# Normalisation de chaque pixel
pixels_normalises = list(map(lambda p: p / 255.0, pixels))

print(f"Pixels normalis√©s : {pixels_normalises}")
````

### B. `filter()` : S√©lection conditionnelle

`filter(fonction, s√©quence)` retourne uniquement les √©l√©ments pour lesquels la fonction (qui doit retourner un bool√©en) est **Vraie**.

### ‚úèÔ∏è Exercice 4.2 (Filter)

1.  Utilisez `filter()` et une expression **lambda** pour s√©lectionner uniquement les scores de confiance qui sont sup√©rieurs √† 0.75.
2.  Convertissez le r√©sultat en liste pour l'affichage.

````python
scores_confiance = [0.92, 0.65, 0.88, 0.70, 0.99, 0.50]

# Filtrage des scores > 0.75
scores_eleves = list(filter(lambda score: score > 0.75, scores_confiance))

print(f"Scores √©lev√©s : {scores_eleves}")
````

### C. `reduce()` : Agr√©gation (N√©cessite `functools`)

`reduce(fonction, s√©quence)` applique la fonction cumulativement aux √©l√©ments de la s√©quence, r√©duisant la s√©quence √† une seule valeur.

### ‚úèÔ∏è Exercice 4.3 (Reduce)

1.  Importez `reduce` depuis le module `functools`.
2.  Utilisez `reduce()` et une expression lambda pour calculer le **produit total** de tous les √©l√©ments de la liste `facteurs`.

````python
from functools import reduce

facteurs = [2, 3, 4, 5]

# Calcul du produit (2 * 3 * 4 * 5)
produit_total = reduce(lambda x, y: x * y, facteurs)

print(f"Produit total : {produit_total}")
````
