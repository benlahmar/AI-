## 2. Attributs et Constructeur (`__init__`)

Le **Constructeur** (`__init__`) est une m√©thode sp√©ciale appel√©e lors de la cr√©ation d'un objet. Elle permet d'initialiser ses **Attributs** (ses donn√©es internes). Le premier argument de toute m√©thode de classe est toujours `self`, qui repr√©sente l'instance elle-m√™me.

### ‚úèÔ∏è Exercice 2.1 (Constructeur et Attributs)

1.  Modifiez la classe `Echantillon` pour inclure un constructeur (`__init__`) qui accepte `id_specimen` et `label`.
2.  Initialisez ces deux attributs en utilisant `self`.
3.  Cr√©ez une nouvelle instance `e2` avec `id_specimen=102` et `label='cat'`.
4.  Affichez la valeur de l'attribut `e2.label`.

```python
class Echantillon:
    # 1. D√©finition du Constructeur
    def __init__(self, id_specimen, label):
        self.id = id_specimen
        self.label = label
        self.est_traite = False # Attribut par d√©faut

# 2. Cr√©ation de l'instance e2
e2 = Echantillon(id_specimen=102, label='cat')

# 3. Affichage d'un attribut
print(f"Label de l'√©chantillon e2 : {e2.label}")
print(f"√âtat initial : {e2.est_traite}")
```

## 3. M√©thodes (Actions)

Les **M√©thodes** sont des fonctions d√©finies √† l'int√©rieur d'une classe. Elles d√©crivent les actions que l'objet peut effectuer ou la mani√®re dont il interagit avec ses attributs.

### ‚úèÔ∏è Exercice 3.1 (D√©finition et Appel de M√©thode)

1.  Ajoutez une m√©thode nomm√©e `marquer_traite` √† la classe `Echantillon`.
2.  Cette m√©thode doit simplement changer l'attribut `self.est_traite` √† `True`.
3.  Cr√©ez une nouvelle instance `e3` (`id=103`, `label='dog'`).
4.  Appelez la m√©thode `e3.marquer_traite()` et v√©rifiez que l'attribut `est_traite` est bien mis √† jour.

```python
class Echantillon:
    def __init__(self, id_specimen, label):
        self.id = id_specimen
        self.label = label
        self.est_traite = False
        
    # 1. D√©finition de la m√©thode
    def marquer_traite(self):
        self.est_traite = True
        print(f"√âchantillon {self.id} marqu√© comme trait√©.")

# 2. Cr√©ation et appel
e3 = Echantillon(id_specimen=103, label='dog')

print(f"Avant traitement : {e3.est_traite}")
e3.marquer_traite()
print(f"Apr√®s traitement : {e3.est_traite}")
```

## 4. Introduction √† l'H√©ritage

L'**H√©ritage** permet √† une nouvelle classe (**Classe Enfant**) d'acqu√©rir les attributs et m√©thodes d'une classe existante (**Classe M√®re**). Cela favorise la r√©utilisation du code.

### ‚úèÔ∏è Exercice 4.1 (H√©ritage Simple)

Nous allons cr√©er une classe `ImageEchantillon` qui h√©rite de `Echantillon`.

1.  D√©finissez la classe `ImageEchantillon` qui h√©rite de `Echantillon`.
2.  Dans son constructeur, appelez le constructeur de la classe m√®re (`super().__init__()`) et ajoutez un nouvel attribut sp√©cifique : `self.chemin_fichier`.
3.  Cr√©ez une instance `img1`.
4.  V√©rifiez que `img1` a h√©rit√© de la m√©thode `marquer_traite` en l'appelant.

```python
# Classe Echantillon r√©utilis√©e (Classe M√®re)
class Echantillon:
    def __init__(self, id_specimen, label):
        self.id = id_specimen
        self.label = label
        self.est_traite = False
        
    def marquer_traite(self):
        self.est_traite = True

# 1. D√©finition de la Classe Enfant
class ImageEchantillon(Echantillon):
    def __init__(self, id_specimen, label, chemin):
        # 2. Appel au constructeur de la classe m√®re
        super().__init__(id_specimen, label)
        self.chemin_fichier = chemin
        
# 3. Cr√©ation de l'instance et test
img1 = ImageEchantillon(104, 'plane', 'data/img_001.jpg')

print(f"Chemin sp√©cifique : {img1.chemin_fichier}")
print(f"√âtat h√©rit√© avant : {img1.est_traite}")

# 4. Utilisation de la m√©thode h√©rit√©e
img1.marquer_traite()
print(f"√âtat h√©rit√© apr√®s : {img1.est_traite}")
```
