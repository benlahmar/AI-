## 1. Organisation du Code : Les Fonctions

Les **fonctions** permettent de r√©utiliser un bloc de code pour effectuer des t√¢ches pr√©cises (ex: calculer une erreur, appliquer une activation).

### ‚úèÔ∏è Exercice 1.1 (Fonction Simple)

1.  D√©finissez une fonction nomm√©e `calculer_erreur` qui prend deux arguments : `valeur_predite` et `valeur_reelle`.
2.  Elle doit retourner la **diff√©rence absolue** entre les deux valeurs (utilisez la fonction `abs()`).
3.  Appelez la fonction avec `valeur_predite = 15` et `valeur_reelle = 18` et affichez le r√©sultat.

```python
def calculer_erreur(valeur_predite, valeur_reelle):
    # Votre code ici
    return abs(valeur_predite - valeur_reelle)

# Appel et affichage
resultat = calculer_erreur(15, 18)
print(f"L'erreur est : {resultat}")
```

## 2. NumPy : Le Langage du Calcul Matriciel

**NumPy (Numeric Python)** fournit des tableaux (`ndarray`) qui sont plus rapides et plus efficaces que les listes Python pour les op√©rations math√©matiques. 

### A. Lancement et Cr√©ation de Tableaux

L'alias standard est `np`.

### ‚úèÔ∏è Exercice 2.1 (Cr√©ation)

1.  Importez la librairie NumPy.
2.  Cr√©ez un tableau 1D (vecteur) √† partir d'une liste Python : `[1, 2, 3, 4, 5]`.
3.  Cr√©ez une matrice 2D (2 lignes, 3 colonnes) avec des z√©ros en utilisant `np.zeros()`.

```python
import numpy as np

# 1. Cr√©ation d'un vecteur
vecteur = np.array([1, 2, 3, 4, 5])
print(f"Vecteur : {vecteur}\n")

# 2. Cr√©ation de la matrice 2x3 de z√©ros
matrice_zeros = np.zeros((2, 3))
print(f"Matrice Z√©ros : \n{matrice_zeros}")
```

### B. Op√©rations Vectorielles (La Magie de NumPy)

Les op√©rations arithm√©tiques sur les tableaux NumPy sont appliqu√©es **√©l√©ment par √©l√©ment**.

### ‚úèÔ∏è Exercice 2.2 (Calculs)

1.  Cr√©ez un tableau `poids` : `[0.1, 0.5, 0.3]`.
2.  Multipliez ce tableau par un scalaire (ex: 2).
3.  Calculez la **somme** de tous les √©l√©ments du tableau `poids` en utilisant la m√©thode **`.sum()`**.

```python
poids = np.array([0.1, 0.5, 0.3])

# 1. Multiplication par un scalaire
poids_doubles = poids * 2
print(f"Poids doubl√©s : {poids_doubles}")

# 2. Somme de tous les poids
somme_poids = poids.sum()
print(f"Somme des poids : {somme_poids}")
```

## 3. Pandas : La Manipulation de Jeux de Donn√©es

**Pandas** est la librairie standard pour manipuler des donn√©es tabulaires. L'objet principal est le **DataFrame**.

### A. Importation et Inspection

L'alias standard est `pd`.

### ‚úèÔ∏è Exercice 3.1 (Cr√©ation et Aper√ßu)

1.  Importez la librairie Pandas.
2.  Cr√©ez un DataFrame √† partir du dictionnaire ci-dessous. 
3.  Affichez les **trois premi√®res lignes** en utilisant la m√©thode **`.head(3)`**.

```python
import pandas as pd

data_patient = {
    'Age': [30, 45, 22, 58, 35],
    'Taux_Cholesterol': [180, 220, 150, 250, 190],
    'Malade': [False, True, False, True, False]
}

df_patients = pd.DataFrame(data_patient)
print("DataFrame complet :\n", df_patients)

# Affichez les trois premi√®res lignes ici
df_patients.head(3)
```

### B. S√©lection et Filtrage de Donn√©es

Les DataFrames permettent de s√©lectionner facilement des colonnes et des lignes selon des crit√®res.

### ‚úèÔ∏è Exercice 3.2 (S√©lection et Filtre)

1.  **S√©lection :** Affichez uniquement la colonne `'Age'`.
2.  **Filtrage :** Cr√©ez un nouveau DataFrame (`df_a_risques`) qui ne contient que les patients ayant un `'Taux_Cholesterol'` sup√©rieur √† **200**.
3.  Affichez ce nouveau DataFrame.

```python
# 1. Afficher uniquement la colonne 'Age'
print("\nColonne Age :\n", df_patients['Age'])

# 2. Cr√©er un filtre pour les taux > 200
filtre = df_patients['Taux_Cholesterol'] > 200

# 3. Appliquer le filtre pour cr√©er le nouveau DataFrame
df_a_risques = df_patients[filtre]

# 4. Afficher le r√©sultat
print("\nPatients √† risques (Chol > 200) :\n", df_a_risques)
```
